  - enrich (left join) fact with mapping tables or static dimension (e.g. dim_date, code, etc)

  - creating new columns. However if the columns created encoded business logic, they should be pushed to the application layer
  - Lookback logic like sessionization, attribution, etc need window functions or fact-fact  joins (depending on how data is modeled) or diff of some kind
